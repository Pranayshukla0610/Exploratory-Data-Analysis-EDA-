```python
import pandas as pd
import numpy as np

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import row, column
from bokeh.models import (
    ColumnDataSource,
    HoverTool,
    Select
)
from bokeh.models.widgets import Div
from bokeh.io import curdoc
```

```python
output_notebook()
```

```python
data = pd.DataFrame({
    'Month':['Jan','Feb','Mar','Apr','May','June'],
    'Sales':[120,150,170,160,180,200],
    'Profit':[10,20,30,40,50,60],
    'Region':['North','South','East','West','North','South']
})
```

```python
source = ColumnDataSource(data)
```

```python
p1 = figure(
    title = 'Monthly Sales',
    x_range = data['Month'],
    width = 600,
    height = 400,
    tools = 'pan,wheel_zoom,box_zoom,reset,save'
)

p1.line(
    x='Month',
    y='Sales',
    source=source,
    line_width=3
)
p1.circle(
    x='Month',
    y='Sales',
    source=source,
    color='red',
    size=10
)

hover = HoverTool(
    tooltips=[
        ('Month','@Month'),
        ('Sales','@Sales'),
        ('Profit','@Profit')
    ]
)

p1.add_tools(hover)
```

```python
p2 = figure(
    title = 'Profit Analysis',
    x_range = data['Month'],
    width=600,
    height=400
)

p2.vbar(
    x = 'Month',
    top = 'Profit',
    source=source,
    width=0.5
)
total_sales = data['Sales'].sum()
total_profit = data['Profit'].sum()

kpi = Div(text=f"""
<h1>Interactive Sales Dashboard</h1>
<h3>Total Sales: {total_sales}</h3>
<h3>Total Profit: {total_profit}</h3>
""")

select = Select(
    title="Select Region",
    value="All",
    options=["All"] + list(data['Region'].unique())
)
```

```python
def update(attr, old, new):

    region = select.value

    if region == "All":
        filtered = data
    else:
        filtered = data[data['Region'] == region]

    source.data = ColumnDataSource(filtered).data

select.on_change('value', update)

layout = column(
    kpi,
    select,
    row(p1, p2)
)
```

```python
show(layout)
```

```python
from bokeh.io import output_notebook, output_file, save, show
output_file("sales_dashboard.html")
```
